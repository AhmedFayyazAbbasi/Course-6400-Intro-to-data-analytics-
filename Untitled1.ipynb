```python
import pandas as pd
from google.colab import files
import io

# Upload a file from your computer
uploaded = files.upload()

# Assuming only one file is uploaded, get its name
for fn in uploaded.keys():
  print(f'User uploaded file "{fn}" with length {len(uploaded[fn])} bytes')
  file_name = fn

# Read the uploaded CSV file into a pandas DataFrame
try:
  df = pd.read_csv(io.BytesIO(uploaded[file_name]))
  print("CSV file loaded successfully into a DataFrame. Here are the first 5 rows:")
  print(df.head())
except Exception as e:
  print(f"Error reading CSV file: {e}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate crime counts by district
crime_counts_by_district = df['DISTRICT'].value_counts().reset_index()
crime_counts_by_district.columns = ['District', 'Crime_Count']

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='District', y='Crime_Count', data=crime_counts_by_district, palette='viridis')
plt.title('Distribution of Crime Counts by District')
plt.xlabel('District')
plt.ylabel('Number of Crimes')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate crime counts by hour
crime_counts_by_hour = df['HOUR'].value_counts().sort_index().reset_index()
crime_counts_by_hour.columns = ['Hour', 'Crime_Count']

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Hour', y='Crime_Count', data=crime_counts_by_hour, palette='viridis')
plt.title('Distribution of Crime Counts by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crimes')
plt.xticks(range(0, 24)) # Ensure all 24 hours are shown on the x-axis
plt.tight_layout()
plt.show()
```
